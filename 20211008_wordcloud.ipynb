```python
import random
```

```python
# 워드클라우드를 그레이톤으로 표시하는 함수를 만든다. => recolor_func 속성에 지정한다.
# word: 워드클라우드로 표시할 단어
# font_size: 워드클라우드로 표시할 단어의 크기
# position: 워드클라우드돌 표시할 단어의 위치
# orientation: 워드클라우드로 표시할 단어의 방향
# hwargs: 워드클라우드로 표시할 단어에 적용되는 글꼴의 경로와 이름
def gray_color_func(word, font_size, position, orientation, random_state=None, **hwargs): # 함수 인수앞에 '*' 가변인자, '**', 넘어오는 데이터 dict로 변환
    #print(word, font_size, position, orientation, **hwargs)
    return 'hsl(0, 0%%, %d%%)' % random.randint(10, 100)
```

```python
plt.figure(figsize=(15, 10))
plt.imshow(wordcloud.recolor(color_func=gray_color_func, random_state=3), interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
import nltk
from konlpy.corpus import kolaw
from konlpy.tag import Twitter
t = Twitter()
```

```python
ko_con_text = kolaw.open('constitution.txt').read()
print(ko_con_text)
```

```python
tokens_ko = t.nouns(ko_con_text)
tokens_ko
```

```python
stop_words = ['제', '년', '월', '일', '조', '수', '때', '그', '이', '바', '및', '정', '관', '위', '장', '거나', '다만'] 
# 형태소 분석결과 (tokens_ko)에서 불용어를 제거하고 다시 저장한다.
# 형태소 분석 결과의 각 단어가 word에 저장되며 반복된다. 이 때, word에 저장된 단어가 stop_words에 포함되지 않는 단어들만
# 리스트로 만들어 형태소 분석 결과로 저장한다.
tokens_ko = [word for word in tokens_ko if word not in stop_words]
tokens_ko
```

```python
# nltk 모듈의 Text 함수의인수를 형태소 분석 결과를 넘겨 단어별 출현 횟수를 계산한다.
ko = nltk.Text(tokens_ko)
ko.vocab().most_common(50)
```

```python
# 단어별 출현 횟수가 많은 순서대로 워드클라우드로 표현할 단어를 추출해서 딕셔너리 형태로 만든다.
data = ko.vocab().most_common(500)
data_wc = dict(data)
data_wc
```

```python
# 딕셔너리에 저장된 데이터는 generate_from_frequencies() 함수를 워드클라우드를 만든다.
wordcloud = WordCloud(
    font_path='C:\\Users\\2020 M\\AppData\\Local\\Microsoft\\Windows\\Fonts\\NanumGothicCoding.ttf',
    background_color='white',
    relative_scaling=0.5,
    mask=mask
).generate_from_frequencies(data_wc)
```

```python
plt.figure(figsize=(15, 10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
